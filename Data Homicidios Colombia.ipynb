```python
import pandas as pd

df = pd.read_csv('/content/HOMICIDIO_20260518.csv')
```

```python
df.columns
```

```python
# Convertir fecha correctamente
df['FECHA HECHO'] = pd.to_datetime(
    df['FECHA HECHO'],
    dayfirst=True,
    errors='coerce'
)

# Crear columna año
df['AÑO'] = df['FECHA HECHO'].dt.year

# Filtrar entre 2017 y 2023
df_filtrado = df[
    (df['AÑO'] >= 2017) &
    (df['AÑO'] <= 2023)
]

# Eliminar la columna AÑO
df_filtrado = df_filtrado.drop(columns=['AÑO'])

# Verificar resultado
print(df_filtrado.head())

# Ver cantidad de registros
print(df_filtrado.shape)
```

```python
df_filtrado.shape
```

```python
df_filtrado.to_csv(
    'Homicidios_Colombia_2017_2023.csv',
    index=False
)
```
